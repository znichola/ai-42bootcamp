```python
# Extract the car's company's name from the variable 'CarName' into the variable 'CarCompany'

CarData['CarCompany'] = CarData['CarName'].map(lambda c: c.split()[0])

CarData.head()
```

```python
# #checking the unique values under "CarCompany"
CarData['CarCompany'].unique()
```

```python
# replacing the mis-spelling with correct ones
def correct_spelling(x: str) -> str:
    corrections = {
        'maxda': 'mazda',
        'Nissan': 'nissan',
        'porcshce': 'porsche',
        'toyouta': 'toyota',
        'vokswagen': 'volkswagen',
        'vw' : 'volkswagen',
    }
    if x in corrections.keys():
        return corrections[x]
    return x

CarData['CarCompany'] = CarData['CarCompany'].map(correct_spelling)

CarData.fuelsystem.replace('mfi','mpfi',inplace=True)

# again checking the unique values
CarData.CarCompany.unique()
```

```python
#creating a dict 'company price' where key = car's company and value = median of their price

company_price = CarData.groupby('CarCompany')['price'].median()
company_price = dict(company_price)

company_price
```

```python
# division in the buckets of low i.e. below 10000, medium i.e. range b/w 10000 and 20000 and high i.e. above 20000

company_price2 = dict(map(lambda v: (v[0], 'low' if v[1] <= 10000 else 'med' if v[1] <= 20000 else 'high'), company_price.items()))

print(company_price2)

# # mapping the company_price on the 'CarCompany' column in the dataset

CarData['CarCompany'] = CarData['CarCompany'].map(company_price2) 

# checking the column's unique values

CarData.CarCompany.unique()
```

```python
CarData = CarData.drop('CarName', axis=1)
CarData = CarData.drop('car_ID', axis=1)
CarData = CarData.drop('symboling', axis=1)
CarData.head()
```

```python
#creating dummy variable for categorical variables
CarData = pd.get_dummies(CarData)
```

```python
#checking the dataset
# print (CarData.info(verbose=True))

CarData.head()
```

```python
# importing required library
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively

CarData_train, CarData_test = train_test_split(CarData, train_size=0.7, test_size=0.3, random_state=42)

# shape of the train and test sets
print(CarData_train.shape)
print(CarData_test.shape)
```

```python
#importing required function for scaling

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python

# y_train = CarData_train['price']

# X_train = CarData_train.drop('price', axis=1)

# scaler.fit(X_train) 
# X_train = scaler.transform(X_train)

# y_test = CarData_test['price']

# X_test = CarData_test.drop('price', axis=1)
# X_test = scaler.transform(X_test)

y_train = CarData_train['price']
X_train = CarData_train.drop('price', axis=1)
X_train = scaler.fit_transform(X_train)

y_test = CarData_test['price']

X_test = CarData_test.drop('price', axis=1)
X_test = scaler.transform(X_test)
```

```python
# Train your model
from sklearn.linear_model import LinearRegression

regression = LinearRegression()

regression.fit(X_train, y_train)
```

```python
# Making predictions
predicted_price = regression.predict(X_test)
```

```python
# Show predictions
predicted_price
```

```python
# Plotting y_test and y_pred to understand the spread.
sns.scatterplot(x=y_test, y=predicted_price)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, predicted_price)
```
