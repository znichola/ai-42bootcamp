```python
# Read the data from the CSV file with read from pandas
# data = pd.read_csv("Salary_dataset.csv")
data = pd.read_csv("./0x0/Salary_dataset.csv")

data = data[["YearsExperience", "Salary"]]
print(data.head(3))

# Configure the style
sns.set_style("darkgrid", {'grid.color': '#b0b0b0'})

g = sns.JointGrid(data=data, x="YearsExperience", y="Salary")
g.plot(sns.scatterplot, sns.histplot)

# Plot the data points with seaborn
# sns.scatterplot(data=data, x="YearsExperience", y="Salary")
```

```python
# We're going to use this library and find out how it works on the Internet
from sklearn.linear_model import LinearRegression

# Separate the features (YearsExperience) from the target variable (Salary)
X = data[["YearsExperience"]]
y = data["Salary"]

# Create a linear regression model
regression = LinearRegression()

# Train the model
train = regression.fit(X, y)

# Predict the Salary values based on YearsExperience
predictions = regression.predict(pd.DataFrame({"YearsExperience" : [10]}))

print(predictions)
```
